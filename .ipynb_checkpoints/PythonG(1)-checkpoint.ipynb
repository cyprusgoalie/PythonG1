For a Gaussian source with spread, it can be represented in Fourier space by: $$\tilde\alpha(\Omega)=\alpha_{c}exp(-2\pi(\nu-\Omega)^{2}/\Delta\nu^{2}$$

and in real space by taking the inverse Fourier Transform: $$\alpha(t)=\frac{1}{2\pi}\int_{-\infty}^{\infty}\tilde\alpha(\Omega)=\alpha_{c}exp(-2\pi(\nu-\Omega)^{2}/\Delta\nu^{2}d\Omega$$
$$\alpha(t)=\alpha_{c}\frac{\Delta\nu}{\sqrt{2}}exp(-\pi\Delta\nu^2t^2)$$

The first order correlation function is: $$g^{(1)}(\tau)=\frac{\langle \alpha^*(t)\alpha(t+\tau) \rangle}{\langle|\alpha_{c}|^{2}\rangle}$$

which gives a Gaussian function with a width inversley proportional to the frequency spread:
$$g^{(1)}(\tau)=exp(-\frac{\pi}{2}\Delta\nu^2\tau^2)$$

Source: Bachor and Ralph, *A Guide to Experiments in Quantum Optics*

For a pulsed laser, the Pulse Rep Rate (PRR) is inversly proportional to the time between pulses (T): $$T=\frac{1}{PRR}$$

For our NKT Photonics SuperK Extreme laser has a variable rep rate ranging from 78MHz-2MHz in 23 steps (approximately 3.3 MHz per step).

At the highest pulse rep rate of 78MHz, the time between pulses is: $$T=\frac{1}{78MHz}=1.282*10^{-8}s=12.82ns$$

The distance travelled by the first pulse in 12.82ns is: $$x = c*12.82ns=3.846m$$

Therefore, if the total travel distance of the initial pulse is below 3.846m, it will never interfere with a following pulse.

For a Michelson Interferometer using a pulsed laser source, multiple pulses may be travelling through the system simultaneously. The number depends on the PRR and the size of the interferometer. If the interferometer's beamsplitter outputs to two legs, a and b, where leg a has a moveable mirror on some mechanical or piezo stage, and leg b is stationary, then the length required for leg a to be for a pulse to interferce with the next incoming pulse is given by:

$$a=\frac{\frac{c}{PRR}+2b}{2}$$

Therefore, assuming the stationary leg's mirror is 0.1 meters from the beamsplitted, the right leg mirror will range a distance of 2.023 meters for 78MHz PRR to 75.1 meters for 2MHz PRR.
